# Solar Radiation Hunts Pt

```python
import rasterio
import numpy as np

# Input file paths
slope_path = r"D:\DSM\slope\slope_huntspt.tif"
solar_rad_path = r"D:\DSM\sun_rad\Hunts_Pt_solar_rad_2025.tif"
output_path = r"D:\DSM\sun_rad\hunts_pt_solar_Rad_S.tif"

output_mask_path = r"D:\DSM\slope\slope_mask_leq_10.tif"

# Threshold slope value
max_slope = 10

# Open the slope raster
with rasterio.open(slope_path) as slope_src:
    # Read the data
    slope = slope_src.read(1)

    # Create the mask: 1 where slope <= max_slope, 0 otherwise
    slope_mask = np.where(slope <= max_slope, 1, 0)

    # Use metadata from the slope raster for the output
    out_meta = slope_src.meta
    out_meta.update(dtype=rasterio.uint8, nodata=0)

    # Save the mask raster
    with rasterio.open(output_mask_path, "w", **out_meta) as dst:
        dst.write(slope_mask.astype(np.uint8), 1)

print("Slope mask raster created as:", output_mask_path)
```

```python
import rasterio
import numpy as np
from matplotlib import pyplot as plt

# File paths
slope_path = r"D:\DSM\slope\slope_huntspt.tif"
solar_rad_path = r"D:\DSM\sun_rad\Hunts_Pt_solar_rad_2025.tif"
output_mask_path = r"D:\DSM\slope\slope_mask_leq_10.tif"

# Plot function to display a raster
def plot_raster(filepath, title, cmap='YlOrRd'):
    with rasterio.open(filepath) as src:
        data = src.read(1)
        # Mask NoData values for visualization
        data = np.ma.masked_equal(data, src.nodata)

        plt.figure(figsize=(8, 8))
        plt.title(title)
        plt.imshow(data, cmap=cmap)
        plt.colorbar(label='Value')
        plt.axis('off')
        plt.show()

# Plot the rasters one by one
plot_raster(slope_path, "Slope Raster", cmap='viridis')
plot_raster(solar_rad_path, "Solar Radiation Raster", cmap='YlOrBr')
plot_raster(output_mask_path, "Slope Mask (<=10)", cmap='gray')
```

```python
import rasterio
import numpy as np

# File paths
slope_mask_path = r"D:\DSM\slope\slope_mask_leq_10.tif"
solar_rad_path = r"D:\DSM\sun_rad\huntspt_solar.tif"
solar_rad_slope_filter_path = r"D:\DSM\sun_rad\solar_rad_slope_filter_1200.tif"

# Open the slope mask raster
with rasterio.open(slope_mask_path) as slope_mask_src:
    slope_mask = slope_mask_src.read(1)

# Open the solar radiation raster
with rasterio.open(solar_rad_path) as solar_src:
    solar_rad = solar_src.read(1)
    solar_meta = solar_src.meta.copy()

# Apply condition: slope_mask == 1 AND solar_rad >= 1200
solar_rad_slope_filter = np.where((slope_mask == 1) & (solar_rad >= 1200), solar_rad, np.nan)

# Update metadata for the filtered raster
solar_meta.update({
    "dtype": "float32",
    "nodata": np.nan
})

# Save the filtered raster
with rasterio.open(solar_rad_slope_filter_path, "w", **solar_meta) as dst:
    dst.write(solar_rad_slope_filter.astype(np.float32), 1)

print("Filtered solar radiation raster with slope <= 10 and solar radiation >= 1200 created as:", solar_rad_slope_filter_path)
```

```python
# import geopandas as gpd
import numpy as np

# Load the rooftop GeoDataFrame


# Initialize suitability as 0
hunts_pt_rooftop_mean['suitability'] = 0

# Apply the first criterion (mean solar radiation)
hunts_pt_rooftop_mean.loc[hunts_pt_rooftop_mean['mean_solar'] >= 529, 'suitability'] = 1

# Second criterion: operation and IntExt
conditions = [
    (hunts_pt_rooftop_mean['operation_'] == 'Large Scale') & (hunts_pt_rooftop_mean['IntExt'] == 'Intensive'),
    (hunts_pt_rooftop_mean['operation_'] == 'Medium Scale') & (hunts_pt_rooftop_mean['IntExt'] == 'Intensive'),
    (hunts_pt_rooftop_mean['operation_'] == 'Large Scale') & (hunts_pt_rooftop_mean['IntExt'] == 'Extensive'),
    (hunts_pt_rooftop_mean['operation_'] == 'Medium Scale') & (hunts_pt_rooftop_mean['IntExt'] == 'Extensive'),
    (hunts_pt_rooftop_mean['operation_'] == 'Small Scale') & (hunts_pt_rooftop_mean['IntExt'] == 'Intensive'),
    (hunts_pt_rooftop_mean['operation_'] == 'Small Scale') & (hunts_pt_rooftop_mean['IntExt'] == 'Extensive')
]

# Corresponding suitability values
values = [6, 5, 4, 3, 2, 1]

# Apply the second criterion to the suitability column
hunts_pt_rooftop_mean['suitability'] = np.select(conditions, values, default=0)

# If mean solar radiation < 1200, reset suitability to 0
hunts_pt_rooftop_mean.loc[hunts_pt_rooftop_mean['mean_solar'] < 529, 'suitability'] = 0

# Save the updated GeoDataFrame
output_path = r"D:\DSM\hunts_pt_rooftop_suitability.shp"
hunts_pt_rooftop.to_file(output_path)

print(f"Updated GeoDataFrame saved to: {output_path}")
```
